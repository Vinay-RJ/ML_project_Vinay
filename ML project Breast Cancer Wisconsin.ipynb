```python
# Checking for outliers
## Z-Score method to remove outliers
from scipy.stats import zscore

df = Raw_data.apply(zscore)
outlier= df[(~(df<3).all(axis=1))].index # Standard Threshold of 3
df = df.drop(outlier)

df
```

```python
# Feature Normalization
### Whole dataset is numeric

from sklearn.preprocessing import MinMaxScaler

x = df.values #returns a numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx [0:2]
```

```python
Norm_X = pd.DataFrame(feature_mtx)
Norm_X.describe()
```

```python
# Reassigning the column names
Norm_X.columns = df.columns
```

## Feature Correlation

```python
## Spearman Correlation plot
# A value of +1 means a perfect association of rank.
# A value of 0 means that there is no association between ranks.
# A value of -1 means a perfect negative association of rank.

import seaborn as sns
df_corr = Norm_X.iloc[:, 0:(Norm_X.shape[1])].corr(method = 'spearman') # Spearman rank correlation

df_corr
```

## Feature Selection

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18,12), dpi=120)         # Sample figsize in inches
ax.tick_params(axis='x', rotation=90)
ax.set_title('Spearman Correlation Plot', fontdict={'fontsize': 2, 'fontweight': 'medium'})

ax = sns.heatmap(df_corr, annot=True, linewidths= 1, ax=ax, fmt=".2f", cmap='Blues')
# ax.set(xlabel="", ylabel="")
ax.xaxis.tick_top()
```

```python
# Hierarchical Clustering and show dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform

plt.figure(figsize=(12,5))

dissimilarity = 1 - abs(df_corr)

Z = linkage(squareform(dissimilarity), 'complete')

dendrogram(Z, labels=Norm_X.columns, orientation='top', 
           leaf_rotation=90)
```

```python
# Clusterize the data and setting the threshold

threshold = 0.4
labels = fcluster(Z, threshold, criterion='distance')

# Show the cluster
labels
```

```python
import numpy as np

# Keep the indices to sort labels
labels_order = np.argsort(labels)

# Build a new dataframe with the sorted columns
for idx, i in enumerate(Norm_X.columns[labels_order]):
    if idx == 0:
        clustered = pd.DataFrame(Norm_X[i])
    else:
        df_to_append = pd.DataFrame(Norm_X[i])
        clustered = pd.concat([clustered, df_to_append], axis=1)
```

## Plot Hierarchical Spearman Correlation Plot

## Random Forest

```python
# evaluate random forest algorithm for classification
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
```

```python
y = y.map({'M': 1, 'B': 0}).astype(int)
```

```python
y.value_counts()
```

```python
# Use numpy to convert to arrays
import numpy as np

# Labels are the values we want to predict
labels = np.array(y)

# Convert to numpy array
features = np.array(Norm_X)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.3, random_state = 50, stratify=labels)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from datetime import datetime as dt
# start = dt.now()

# running_secs = (dt.now() - start).seconds
```

```python
from matplotlib import pyplot

start = dt.now()

# define the model
model_rf = RandomForestClassifier()
model_rf.fit(train_features, train_labels)
test_labels_pred=model_rf.predict(test_features)

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=7, n_repeats=3, random_state= 50) # Default hyperparameters
n_scores = cross_val_score(model_rf, train_features, train_labels, scoring='accuracy', cv= cv, n_jobs=-1, error_score='raise')

running_secs = (dt.now() - start).seconds

print('\033[1mRandom Forest Run Time', running_secs, '\033[0m')

print('\033[1mAverage performance Score across all folds:',n_scores.mean(),'\033[0m')


# get importance
importance = model_rf.feature_importances_

# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))

# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
# Report Model Performance metrics

# print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
# print('$$$$$$$$$$$$$$$$$$$$$$$$')
print("Accuracy:\033[1m",metrics.accuracy_score(test_labels, test_labels_pred),"\033[0m")
print("confusion_matrix : \n", confusion_matrix(test_labels, test_labels_pred))
print("classification_report : \n", classification_report(test_labels, test_labels_pred))

cf_matrix_rf = confusion_matrix(test_labels, test_labels_pred)
```

```python
# Confusion Matirx

import seaborn as sns
sns.set(font_scale=2.0)
sns.heatmap(cf_matrix_rf, annot= True, cmap='Blues')
```

```python
# AUC curve

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)

test_labels_pred_proba = model_rf.predict_proba(test_features)[::,1]
fpr, tpr, _ = metrics.roc_curve(test_labels,  test_labels_pred_proba)
auc = metrics.roc_auc_score(test_labels, test_labels_pred_proba)
plt.plot(fpr,tpr,label="AUC="+str(auc).format(auc))

plt.plot([0,1],[0,1],'k--')
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC Curve')

plt.legend(loc = 'lower right', prop={'size': 20})
plt.show()
```
